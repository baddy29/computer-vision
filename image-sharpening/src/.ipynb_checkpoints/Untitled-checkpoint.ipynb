```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read test images
cat_img = cv2.imread('cat_crp.jpg')
#Read Dog Image
dog_img = cv2.imread('dog_crp.jpg')
```

```python
#Kernel and sigma list for experimentation
kernel_sizes=[(71,71),(19,19)]
sigmas=[99,17,3]
```

```python
#Gaussian blur and sharpening on cat image
cat_low_pass_filter_list={}
for kernel in kernel_sizes:
    for sigma in sigmas:
        gauss_blur_cat=cv2.GaussianBlur(cat_img,kernel,sigma)
        cat_blur_kernel=f'cat_blur_kernel_{kernel[0]}_sigma_{sigma}'
        cat_low_pass_filter_list[cat_blur_kernel]=gauss_blur_cat
        
        low_pass_filter_cat_file=f'experiment_results/Cat_Images/low_filter/{cat_blur_kernel}.jpg'
        cv2.imwrite(low_pass_filter_cat_file,gauss_blur_cat)
        
        img_diff_cat=np.subtract(cat_img.astype(int),gauss_blur_cat.astype(int))
        sharp_img_cat=np.add(cat_img.astype(int),img_diff_cat.astype(int))
        #imwrite() will take care or nomalization of image
        sharpened_img_output=f'experiment_results/Cat_Images/sharpened/cat_sharpened_k{kernel[0]}s{sigma}.jpg'
        cv2.imwrite(sharpened_img_output,sharp_img_cat)
```

```python
#Gaussian blur and sharpening on cat image
dog_high_pass_filter_list={}
for kernel in kernel_sizes:
    for sigma in sigmas:
        gauss_blur_dog=cv2.GaussianBlur(dog_img,kernel,sigma)
        dog_blur_kernel=f'dog_blur_kernel_{kernel[0]}_sigma_{sigma}'
        
        low_pass_filter_dog_file=f'experiment_results/Dog_Images/low_filter/{dog_blur_kernel}.jpg'
        cv2.imwrite(low_pass_filter_dog_file,gauss_blur_dog)
        
        #High pass filter for dog
        img_diff_dog=np.subtract(dog_img.astype(int),gauss_blur_dog.astype(int))
        dog_high_pass_filter_list[dog_blur_kernel]=img_diff_dog
        
        sharp_img_dog=np.add(dog_img.astype(int),img_diff_dog.astype(int))
        #imwrite() will take care or nomalization of image
        sharpened_img_output=f'experiment_results/Dog_Images/sharpened/dog_sharpened_k{kernel[0]}s{sigma}.jpg'
        cv2.imwrite(sharpened_img_output,sharp_img_dog)
```

```python
filter_cat71_99=cv2.imread('experiment_results/Cat_Images/low_filter/cat_blur_kernel_71_sigma_99.jpg')
sharpened_cat71_99=cv2.imread('experiment_results/Cat_Images/sharpened/cat_sharpened_k71s99.jpg')
```

```python
#Split Original and Sharpened into RGB and Comparing histograms
images = [cat_img,filter_cat71_99,sharpened_cat71_99]
titles = ['Original Cat','Filter 71_99','Sharpen 71_99']

plt.figure(figsize=(15, 10))

for i, img in enumerate(images, 1):
    # Split the image into its RGB channels
    blue, green, red = cv2.split(img)

    plt.subplot(5, 3, i * 3 - 2)
    plt.hist(blue.flatten(), bins=256, color='blue', range=[0, 256])
    plt.title(f'{titles[i-1]} Blue Channel Histogram')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')

    plt.subplot(5, 3, i * 3 - 1)
    plt.hist(green.flatten(), bins=256, color='green', range=[0, 256])
    plt.title(f'{titles[i-1]} Green Channel Histogram')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')

    plt.subplot(5, 3, i * 3)
    plt.hist(red.flatten(), bins=256, color='red', range=[0, 256])
    plt.title(f'{titles[i-1]} Red Channel Histogram')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

def hybrid_image(cat_blur, dog_high_pass,output_path):
#     print(dog_high_pass_filter_list[dog_high_pass])
    hybrid_result = np.add(cat_blur.astype(int), dog_high_pass)
    cv2.imwrite(output_path, hybrid_result)
    
def create_and_save_hybrid(cat_low_pass_filter_list, dog_high_pass_filter_list):
    i=0
    j=0
    filter_name_keys1=list(dog_high_pass_filter_list.keys())
    filter_name_keys2=list(cat_low_pass_filter_list.keys())
    for dog_high_pass in dog_high_pass_filter_list:
        key_name1=filter_name_keys1[i]
        key_split1=key_name1.split('_')
        hybrid_filter_key1=key_split1[3]+key_split1[5]
        for cat_blur in cat_low_pass_filter_list:
            key_name2=filter_name_keys2[j]
            key_split2=key_name2.split('_')
            hybrid_filter_key2=key_split2[3]+key_split2[5]
            output_path = f'experiment_results/hybrid/hybrid_cat_dog_{hybrid_filter_key2}_{hybrid_filter_key1}.jpg'
            hybrid_image(cat_low_pass_filter_list[key_name2], dog_high_pass_filter_list[key_name1],output_path)
            j+=1
        j=0
        i+=1

```

```python
# Create and save hybrid images
create_and_save_hybrid(cat_low_pass_filter_list, dog_high_pass_filter_list)
```
